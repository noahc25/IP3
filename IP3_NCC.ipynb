## Remove all the symbols from the 'Price' column and convert  and update it to  64-bit int numbers using astype(‘int64’) Hint. Look for string functions like replace/strip).

```python
data['Price']=data['Price'].str.replace(',','', regex=True).str.replace('$','', regex=True)
```

```python
data['Price']=data['Price'].astype('int64')
```

```python
data.info()
```

## Add a column called 'Total Sales' to car_sales which cumulatively adds the 'Price' column look up the cumsum() function.


```python
data['Total Sales']=np.cumsum(data['Price'])
data
```

## Add a column called 'Sale Date' which lists a series of successive dates starting from today (your today). Check out “date_range” function and keep the periods=len(car_sales).

## Use the matplotlib line plot function to plot the 'Price' column versus the 'Total Sales' column (Give proper title and appropriate labels.


```python
fig=plt.figure(figsize=(10,6))
ax=fig.add_axes([0,0,1,1])
ax.set_title('Price vs Total Sales')
plt.xlabel('Price')
plt.ylabel('Total Sales')
ax.plot(data['Price'], data['Total Sales'], color='cornflowerblue')
plt.show()
```

## Create a matplotlib scatter plot of the 'Odometer (KM)' vs 'Price' column.

```python
plt.scatter(data['Odometer (KM)'], data['Price'])
ax1=fig.add_axes([1,1,2,2])
plt.title('Odometer (KM) vs Price')
plt.xlabel('Odometer (KM)')
plt.ylabel('Price')
plt.show()
```

## Find out Pearson correlation between 'Total Sales','Price','Odometer (KM)'. Comment on the correlation
